### 17. Is there any relationship between reviews and price?
From the observations below, airbnbs which having lower reviews approching 0, will have ralatively large price viarations.
In addition, people loves to review airbnbs with relatively lower  prive and which with lower amount of price competitors

```python
air[['reviews','price']].boxplot(by='reviews',figsize=(14,6))
plt.title("The potential relationship between reviews and the price 1",fontsize=22)
plt.show()
```

```python
avgprice=air.groupby(by=['reviews'],dropna=False)['price'].mean()
plt.figure(figsize=(14,6))
plt.title("The potential relationship between reviews and the price 2",fontsize=22)
plt.plot(avgprice)
plt.xlabel("# of reviews",fontsize=15)
plt.ylabel('The average prices group by reviews',fontsize=15)
plt.show()
```

### 18. Is there any relationship between satisfiction and price?
It is interstingly to realize that the overall_satisfaction rating does not have a linear relationship with the price.
The highest price range occurs when the satisfactions lay between 2-3, and for airbnbs with low satisfaction or high satisfiction, they
have relatively lower average prices.

```python
avgprice=air.groupby(by=['overall_satisfaction'],dropna=False)['price'].mean()
plt.figure(figsize=(14,6))
plt.title("The potential relationship between satisfiction and the price",fontsize=22)
plt.plot(avgprice)
plt.xlabel("The overall_satisfaction (0-5)",fontsize=15)
plt.ylabel('The average prices group by overall_satisfiction',fontsize=15)
plt.show()
```

### 19. Is there any relationship between number of bedrooms and price?
The prices go up as the bedroom numbers increase, up to around 7 and when the bedroom number is 7, the paverage price is the highest. When the airbnbs have too much bedrooms, greater then 7,
the prices derease but still maintain in a high level.

```python
avgprice=air.groupby(by=['bedrooms'],dropna=False)['price'].mean()
plt.figure(figsize=(14,6))
plt.title("The potential relationship between bedrooms and the price",fontsize=22)
plt.plot(avgprice)
plt.xlabel("The overall_satisfaction (0-5)",fontsize=15)
plt.ylabel('The average prices group by bedrooms',fontsize=15)
plt.show()
```
